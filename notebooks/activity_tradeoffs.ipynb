# Finalized Crp Tradeoff Plot

```python
ica_data.sample_table[ica_data.sample_table['project'] == 'crp']
```

```python
_, ax = plt.subplots(figsize=(5, 4))

color_dict = {
    'wt_fru': 'purple',
    'wt_glyc': 'violet',
    'delar1_glyc': 'tab:blue',
    'delar1ar2_glyc': 'tab:green',
    'delar2_glyc': 'tab:orange',
    'delcrp_fru': 'salmon',
    'delcrp_glc': 'red',
    'delcrp_glyc': 'darkred'
}

label_dict = {
    'wt_fru': 'WT (Fru)',
    'wt_glyc': 'WT (Glyc)',
    'delar1_glyc': 'Ar1 (Glyc)',
    'delar1ar2_glyc': 'Ar2Ar2 (Glyc)',
    'delar2_glyc': 'Ar2 (Glyc)',
    'delcrp_fru': 'CRP (Fru)',
    'delcrp_glc': 'CRP (Glc)',
    'delcrp_glyc': 'CRP (Glyc)'
}

for (proj, cond), proj_cond_df in ica_data.sample_table.groupby(['project', 'condition']):
    proj_cond_idx = proj_cond_df.index
    proj_cond_act = ica_data.A[proj_cond_idx].mean(axis=1)
    crp1_act, crp2_act = proj_cond_act['Crp-1'], proj_cond_act['Crp-2']
    if proj == 'crp':
        ax.scatter(crp1_act, crp2_act, s=40, c=color_dict[cond], label=label_dict[cond], zorder=10)
    else:
        ax.scatter(crp1_act, crp2_act, s=15, c='gray')

ax.tick_params(axis='both', labelsize=12)
ax.axvline(0, color='black', linestyle='--', linewidth=0.5)
ax.axhline(0, color='black', linestyle='--', linewidth=0.5)
ax.set_xlabel('Crp-1 iModulon Activity', fontsize=13)
ax.set_ylabel('Crp-2 iModulon Activity', fontsize=13)
ax.legend(fontsize=13, bbox_to_anchor=(1, 0.5), loc='center left')
for s in ax.spines.values():
    s.set_linewidth(1.5)
    
plt.savefig('fig5/crp_tradeoff.svg')
```
